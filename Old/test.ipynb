```python
from uta_infer import uta_rank
import numpy as np

import numpy as np
np.random.seed(42)
# 200 alternatives avec des floats entre 0 et 500 sur 4 critères
X = np.random.rand(20, 4)
print(X[0:5])
```

```python
order = np.arange(X.shape[0])
np.random.shuffle(order)

res = uta_rank(
    X=X,
    L=5,                     # 5 segments pour *chaque* critère
    order=order,          # ordre donné
    scale="minmax",          # standardisation → [0,100]
    pairs="consecutive",     # contraintes sur pairs successifs
    return_details=True      # pour voir les poids, fonctions, etc.
)

print("Ranking inféré (best→worst):\n", res["ranking"])
print("Poids inférés:", np.round(res["weights"], 4))
```

```python
print("Slack :", res["obj_slack"])
```
